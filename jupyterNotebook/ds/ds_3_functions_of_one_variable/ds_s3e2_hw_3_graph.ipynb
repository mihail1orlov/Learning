```python
from sympy import *
from sympy.plotting import plot
```

```python
init_printing(use_unicode=False, wrap_line=False, no_global=True)
```

```python
x = Symbol('x')
```

```python
f = 2*x**4 + x**3 - 6*x**2 + 7*x + 1
f
```

```python
f.subs(x, 3)
```

```python
f.subs(x, -5)
```

```python
plot(f)
```
