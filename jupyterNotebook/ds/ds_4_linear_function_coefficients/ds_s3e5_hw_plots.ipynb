```python
from sympy import *
# imports all basic functions
# sqrt
# ln
from sympy.plotting import plot
```

```python
init_printing(use_unicode=False, wrap_line=False, no_global=True)
```

```python
x = Symbol('x')
```

```python
f1 = x**2-3*x-20
plot(f1)
```

```python
f2 = x**2-3*x+5
plot(f2)
```

```python
f3 = -x**2-3*x+5
plot(f3)
```

```python
f4 = -x**2-3*x-4
plot(f4)
```
