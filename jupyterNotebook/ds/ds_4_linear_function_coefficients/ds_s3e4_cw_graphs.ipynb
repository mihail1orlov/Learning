```python
from sympy import *
# imports all basic functions
# sqrt
# ln
from sympy.plotting import plot
```

```python
init_printing(use_unicode=False, wrap_line=False, no_global=True)
```

```python
x = Symbol('x')
```

```python
f = x**2
f
```

```python
f1 = x**2+10
f1
```

```python
# можно иследовать функции на парность
f.subs(x, 3)
```

```python
f.subs(x, -3)
```

```python
plot(f, f1)
```

```python
f2 = sqrt(x)
f2
```

```python
f2.subs(x, -1)
# We get a complex number
```

```python
plot(f2)
```

```python
f3 = 2*x**2 + 10*x + 50
f3
```

```python
f4 = -2*x**2 + 10*x - 50
f4
```

```python
plot(f3, f4)
```

```python
# Allows you to set the required range of values
plot(f3, f4, (x, -20, 20))
```

```python
f5 = 3*x - 2*x**2
f5
```

```python
f6 = -3*x - 2*x**2
f6
```

```python
plot(f5, f6)
```

```python
f7 = exp(x)
f7
plot(f7)
plot(f7, (x, -1, 4))
```

```python
f8 = ln(x)
f8
plot(f8)
```
