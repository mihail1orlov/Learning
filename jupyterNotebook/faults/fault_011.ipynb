```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Задайте параметры A, B и C
A = 0.577
B = 1
C = 1617
```

```python
# Создайте диапазон значений x
x_values = np.linspace(10000, 20000, 1000)  # 1000 точек от 10000 до 20000
```

```python
# Вычислите соответствующие значения y
y_values = (C - A * x_values) / B
```

```python
# Проверьте неравенство A * x + B * y >= C
result = A * x_values + B * y_values >= C
```

```python
# Постройте график
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label='A*x + B*y = C')
plt.scatter(x_values[result], y_values[result], color='red', label='A*x + B*y >= C')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
