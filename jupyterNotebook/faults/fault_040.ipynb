```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# coefficients
A = 0.9395248110663589
B = -0.342035879551065
C = 0.01745240643728344
D = -5449581.898706565
#"{\"A\":0.9395248110663589,\"B\":-0.342035879551065,\"C\":0.01745240643728344,\"D\":-5449581.898706565}"


# create a grid of points
x = np.linspace(-100,100,10)
y = np.linspace(-100,100,10)
x, y = np.meshgrid(x,y)

# calculate the z coordinates for each point on the plane
z = (D - A*x - B*y) / C

# create 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z)

# rotate the plot
ax.view_init(20, 80)

plt.show()
```
