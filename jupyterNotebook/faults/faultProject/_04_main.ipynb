```python
import matplotlib.pyplot as plt
import numpy as np
from geometry import Line, Point2D
from visualization import visualize

def define_points():
    points = [
        #Point2D(9999.000000001313, 10199.8134765625),
        #Point2D(10002.000000001324, 7486.134356717739),
        Point2D(9000, 8000, color='y', name='yy'),
        Point2D(10000, 9000, color='c', name='cc'),
        Point2D(8000, 8500, color='r', name='rr'),
        Point2D(8000, 8000, color='k', name='kk'),
        Point2D(9000, 9000, color='g', name='gg'),
        Point2D(7000, 9500, color='y', name='y_false'),
        Point2D(10500, 8500, color='m', name='m_true'),
        #Point2D(10002.000000001324, 10199.5546875)
    ]
    return points    

# Initialize the line

points = define_points()

line = Line.from_two_points(points[4], points[3]) # A = 1000, B = -1000, C = 1000000, 45.0 degrees
line.name = 'green'
line.color = 'g'

line_r = Line.from_angle_point(-10, points[2])
line_r.name = 'red'
line_r.color = 'r'

line_y = Line.from_angle_point(45, points[0])
line_y.name = 'yellow'
line_y.color = 'y'

visualize(points, [line, line_r, line_y])
```
