```python
import matplotlib.pyplot as plt
import numpy as np

angle = 91
x = 14017.00000001593
y = 9587.966869618005

if abs(angle - 90) < 0.01 or abs(angle + 90) < 0.01:
    A = 1
    B = 0
    C = x
else:
    # переводим угол из градусов в радианы
    angle_rad = np.deg2rad(angle)
    
    A = np.tan(angle_rad)
    B = 1
    C = y - A * x

#A = -57.28996163075955
#B = 1.0
#C = 812621.3590488873

# Генерация значений x
x = np.linspace(-10000, -20000, 400)  # Вы можете изменить диапазон и количество точек в соответствии с вашими потребностями

# Вычисление значений y
y = (C - A * x) / B

# Построение графика
plt.figure(figsize=(10,6))
plt.plot(x, y, label='A*x + B*y = C')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of the Line')
plt.legend()
plt.grid(True)
plt.show()
```
