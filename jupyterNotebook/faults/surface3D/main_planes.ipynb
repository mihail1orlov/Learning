```python
import numpy as np
from lib.Point3d import Point3d
from lib.Plane import Plane
from lib.PlaneCreator import PlaneCreator
from lib.PlaneAnalyzer import PlaneAnalyzer
from lib.Surface import Surface
from lib.Plotter3d import Plotter3d

minEast = 0
minNorth = 0
maxEast = 1000
maxNorth = 1000
numRows = 10
numCols = 10

def create_surface_with_random_values(minEast=minEast, minNorth=minNorth, maxEast=maxEast, maxNorth=maxNorth, numRows=numRows, numCols=numCols):
    surface = Surface(minEast, minNorth, maxEast, maxNorth, numRows, numCols)
    for row in range(surface.numRows):
        for col in range(surface.numCols):
            surface.set_value(row, col, np.random.rand()/3)
    return surface

plotter = Plotter3d()

points = [Point3d(0, 500, 0), Point3d(1000, 500, 2, 'g'), Point3d(1000, 500, 0, 'm')]
plotter.plot_points(points)
#plotter.plot_points([Point3d(0, 500, 1.5, 'b'), Point3d(1000, 500, 1.5, 'b'), Point3d(1000, 500, 0, 'b')])

plane = PlaneCreator.from_points(points[0], points[1], points[2])
PlaneAnalyzer.print_info(plane)

#plotter.plot_plane(plane, minEast=minEast - 1, maxEast=maxEast + 1, minNorth=minNorth - 1, maxNorth=maxNorth + 1)

loaded_surface = Surface.load_from_file('data/surface_-90.json')
plotter.plot_surface(loaded_surface)

plotter.show_plot()
```
