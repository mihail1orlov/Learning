```python
from lib.Point3d import Point3d
from lib.PointInfoPrinter import PointInfoPrinter

from lib.Plane import Plane
from lib.PlaneAnalyzer import PlaneAnalyzer
from lib.PlaneCreator import PlaneCreator

from lib.Plotter3d import Plotter3d

from lib.SurfacePlotter import SurfacePlotter
from lib.SurfaceIO import SurfaceIO
from lib.SurfaceManager import SurfaceManager

init_points = [Point3d(1200, 300, 1, 'y'), Point3d(1200, 300, 0, 'k'), Point3d(1200, 200, 0, 'm')]
input = 'data/FaultOperationsTests/input.json'
output = 'data/FaultOperationsTests/expectedSplitSurfaceGridByFault_SplitAcrossEastAxis_MiddleRowIsBoundary.json'


plotter_points = Plotter3d()
PointInfoPrinter.print_points(init_points)
plotter_points.plot_points(init_points)
plotter_points.show_plot('input points')

plotter_input = Plotter3d()
surface_input = SurfaceIO.load_from_file(input)
plotter_input.plot_surface(surface_input)
plane = PlaneCreator.from_points(init_points[0], init_points[1], init_points[2])
PlaneAnalyzer.print_info(plane)
points_input = SurfaceManager.process_surface(surface_input, plane)
plotter_input.plot_points(points_input)
plotter_input.show_plot('input')
PointInfoPrinter.print_points(points_input)

plotter_output = Plotter3d()
surface_output = SurfaceIO.load_from_file(output)
points_output = SurfaceManager.process_surface(surface_output, plane)
plotter_output.plot_points(points_output)
plotter_output.plot_surface(surface_output)
plotter_output.show_plot('output')

surfacePlotter = SurfacePlotter()
surfacePlotter.plot_surfaces(surface_input, surface_output, 'input', 'output')

PointInfoPrinter.print_points(points_output)
PointInfoPrinter.print_summary(points_output)
```
