```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from lib.plane import plane

# создаем объект Plane
p = plane(1, 1, 1, 0, 0, 0)

# создаем диапазон значений для x и y
x = np.linspace(-5, 5, 10)
y = np.linspace(-5, 5, 10)

# создаем сетку значений для x и y
X, Y = np.meshgrid(x, y)

# вычисляем соответствующие значения z
Z = (p.D - p.A * X - p.B * Y) / p.C

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# создаем поверхность
ax.plot_surface(X, Y, Z)

# добавляем подписи к осям
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# отображаем график
plt.show()
```
