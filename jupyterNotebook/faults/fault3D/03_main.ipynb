```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from lib.plane import plane

# создаем объект Plane
#
p = plane(0.8137843159659884, -0.2962348512376092, 0.49999999999999994, 1687360.6451473145, 20568102.07540489, 7485.074099872728)

# создаем диапазон значений для x и y
x = np.linspace(-5, 5, 10)
y = np.linspace(-5, 5, 10)

# создаем сетку значений для x и y
X, Y = np.meshgrid(x, y)

# вычисляем соответствующие значения z
Z = (p.D - p.A * X - p.B * Y) / p.C

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# создаем поверхность
ax.plot_surface(X, Y, Z)

# добавляем подписи к осям
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# отображаем график
plt.show()

print(f'plane: {p.D}') # 0.49999999999999994
```
