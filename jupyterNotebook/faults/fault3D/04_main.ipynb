```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from lib.plane import plane

# Создаем объект Plane и устанавливаем свойства
p = plane(1, 1, 1, 0, 0, 0)
p.set_props(2, 3, -1, 0)  # Set the properties of the 'p' instance

# Создаем диапазон значений для x и y
x = np.linspace(-5, 5, 10)
y = np.linspace(-5, 5, 10)

# Создаем сетку значений для x и y
X, Y = np.meshgrid(x, y)

# Вычисляем соответствующие значения z
Z = (p.D - p.A * X - p.B * Y) / p.C

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Создаем поверхность
ax.plot_surface(X, Y, Z, alpha=0.5, rstride=100, cstride=100)

# Добавляем массив точек
points = np.random.rand(10, 3) * 10 - 5  # 10 точек в 3D
ax.scatter(*points.T, color='red')  # Каждая точка будет красного цвета

# Добавляем линию
line_points = np.array([[-5, -5, -5], [5, 5, 5]])  # Две точки, которые определяют линию
ax.plot(*line_points.T, color='blue')  # Линия будет синего цвета

# Добавляем подписи к осям
ax.set_xlabel('East')
ax.set_ylabel('North')
ax.set_zlabel('TVD')

# Отображаем график
plt.show()
```
