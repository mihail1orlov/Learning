```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from lib.line3d import line3d

# Определение точки
x1 = 1
y1 = 1
z1 = 1

# Определение линии
x2 = np.linspace(5, 5, 100)
y2 = x2
z2 = x2

# Определение плоскости
x3 = np.linspace(-5, 5, 100)
y3 = np.linspace(-5, 5, 100)
x3, y3 = np.meshgrid(x3, y3)
z3 = x3 + y3


# Создание нового окна для графика и добавление 3D подграфика
fig = plt.figure(figsize=(16, 16))
ax = fig.add_subplot(111, projection='3d')

# Рисуем точку
ax.scatter(x1, y1, z1, color='r', s=100)

# Рисуем линию
ax.plot(x2, y2, z2, color='b')

# Рисуем плоскость
ax.plot_surface(x3, y3, z3, color='b', alpha=0.5)

#ax.view_init(0, -90)
#ax.view_init(90, -90) # x & y

# Настройка отображения
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```
