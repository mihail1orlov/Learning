```python
import cmath

def solve_quadratic(a, b, c):
    # Вычисляем дискриминант
    D = cmath.sqrt(b**2 - 4*a*c)

    # Находим корни уравнения
    x1 = (-b + D) / (2*a)
    x2 = (-b - D) / (2*a)

    return x1, x2

# Пример использования:
a = 2
b = -18
c = -72

x1, x2 = solve_quadratic(a, b, c)
print(f"Корни квадратного уравнения: x1 = {x1}, x2 = {x2}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def quadratic_function(a, b, c, x):
    return a*x**2 + b*x + c

# График уравнения
margin = 2  # Запас пространства вокруг корней на графике
x_range = np.linspace(min(x1.real, x2.real) - margin, max(x1.real, x2.real) + margin, 400)
y_range = quadratic_function(a, b, c, x_range)

plt.plot(x_range, y_range, label=f'{a}x^2 + {b}x + {c}')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title("График квадратного уравнения")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.xlim([x_range[0], x_range[-1]])
plt.ylim([min(y_range), max(y_range)])
plt.legend()
plt.show()

print(f"Корни квадратного уравнения: x1 = {x1}, x2 = {x2}")
```
